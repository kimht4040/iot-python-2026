```python
import cv2
import numpy as np

cat = cv2.imread(filename="cat.jpeg", flags= cv2.IMREAD_COLOR_BGR)
(b,g,r) = cv2.split(cat)
blue = cv2.merge((b, np.zeros_like(g), np.zeros_like(r)))
green = cv2.merge((np.zeros_like(b),g,np.zeros_like(r)))

cv2.imshow('origin', cat)
cv2.imshow('blue', blue)
cv2.imshow('green', green)
cv2.waitKey(0) # 키 입력 대기
cv2.destroyAllWindows()
for i in range(1, 5):
    cv2.waitKey(1)
```

```python
import cv2
import numpy as np
cat = cv2.imread(filename='cat.jpeg')

cv2.circle(cat, (416,210), 100, (0,255,0), thickness=5, lineType=cv2.LINE_AA)

cv2.imshow('origin', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
for i in range(1, 5):
    cv2.waitKey(1)
```

### ROI
- Region of interest : 관심영역
  - 2차원 리스트(매트릭스 , 행렬)
  - mat[y시작점:y시작점+높이,x시작점:x시작점+넓이]





```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
cat = cv2.imread(filename='cat.jpeg')
roi = cat[81:81+214,295:295+240]
plt.figure(figsize=(5, 4)) # 이미지 크게 보기
plt.imshow(cv2.cvtColor(cat, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
cat = cv2.imread(filename='cat.jpeg')
gray = cv2.cvtColor(cat,cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(gray, threshold1=100, threshold2=200)
# 1번 칸: Gray 이미지
plt.subplot(1, 2, 1) # 1행 2열 중 1번째
plt.imshow(gray, cmap='gray') # 그레이스케일은 cmap='gray'를 쓰는 게 정확합니다
plt.title('Gray Image')
plt.axis('off')

# 2번 칸: Canny 이미지
plt.subplot(1, 2, 2) # 1행 2열 중 2번째
plt.imshow(canny, cmap='gray')
plt.title('Canny Edge')
plt.axis('off')

plt.show()
```
