# 신경망 핵심원리
## 자동미분
- 딥러닝의 심장인 기능
- 학습과정에서 수많은 데이터를 통해서 정답에 점점가까워짐
- 가장 낮은 지점(최소 손실) 찾으려면, 방향과 경사를 수학적으로 계산해야 함 -> 미분
- 기울기 - 데이터들로 판별된 경사도
  - 파라미터 - 이차 방정식 x^2, y에 포함되는 수 
  - 파라미터(편향) - 이차 방정식 마지막 숫자값 
- 손실 - 실제 경사 값과 구해진 경삭닶 사이의 차이 
  - 이 손실을 최소화 해야 알맞은 값이 나옴 
  - 가중치와 편향을 조정하면서 손실이 최소화 되게 학습
- 경사하강법 - 기울기를 조정해서 최소 손실에 가깝게 만드는 과정 

### 자동미분의 역할 
- Autograd : 파이토치가 제공하는 자동미분 엔진 사용자가 텐서에서 어떤 연산을 수행하든 그 과정을 모두 기록 기록을 바탕으로 미분값을 자동 계산
  - requires_grad = true



```python
import torch
```

```python
# 텐서 x의 미분값을 계산하도록 설정
x = torch.tensor(2.0, requires_grad = True)
```
