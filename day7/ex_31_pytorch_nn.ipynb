```python
import torch
import torch.nn as nn #신경망 모듈
import torch.optim as optim #손실함수 최적화알고리즘 모듈

```

```python
w_true = 2.0
b_true = 1.0
#랜덤값 x생성
x = torch.randn(100) *10

#노이즈 생성
noise = torch.randn(100) *2

#y값 생성

y = w_true * x + b_true + noise

#모델설계
class LinearRegression(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__()
        self.w = nn.Parameter(torch.randn(1))
        self.b = nn.Parameter(torch.randn(1))
    def forward(self, x):
        #선형회귀 공식 y = wb + b
        y_pred = self.w * x + self.b
        return y_pred


# 모델 객체 생성
model = LinearRegression()

# 3.손실함수 
# MSE - 예측값과 실제값 차이를 제곱하고 평균
loss_fn = nn.MSELoss()


# 4. 옵티마이저
# SGD(확률적 경사 하강법)


optimizer = optim.SGD(model.parameters(), lr = 0.01)


# 5.모델학습 
num_epochs = 1000

for epoch in range(num_epochs+1):
    #순전파
    y_pred = model(x)
    
    #손실 계산
    loss = loss_fn(y_pred, y)
    
    #역전파
    optimizer.zero_grad() #기울기 초기화
    loss.backward() #기울기 계산
    
    #매개변수 업데이트
    optimizer.step()
    
    if epoch % 100 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item():.4f}, w: {model.w.item():.4f}, b: {model.b.item():.4f}')



print(f'최종 모델 파라미터: w={model.w.item():.4f}, b={model.b.item():.4f}')
```
