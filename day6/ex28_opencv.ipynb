## fps 표시

```python
import time
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)
prev_time = 0


while True:
    ret, img = cap.read()
    if not ret: break
    curr_time = time.time() # 프레임 출력시간
    #fps 계산
    if prev_time != 0 :
        fps = 1/(curr_time - prev_time)
    else:
        fps = 0
    prev_time = curr_time
    #출력할 텍스트
    fps_text = f'FPS : {int(fps)}'
     # 이미지 위에 표시
    cv2.putText(img,fps_text,(100,100), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,255,0), 2)
    cv2.imshow('webcam', img)
   
    if cv2.waitKey(10) == ord('q'):break

cap.release() # 메모리 정리
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
import time  # 추가: time 모듈 임포트

# haar cascade 파일로드
# 현재 설치된 OpenCV의 공식 데이터를 가져옵니다.
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')

# 로드 확인
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise RuntimeError('웹캠 오픈 실패')

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

prev_time = 0
show_info = False 

# 실제 해상도 읽기
real_w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
real_h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

while True:
    ret, img = cap.read()
    if not ret: break
    curr_time = time.time()

    if prev_time != 0 :
        fps = 1/(curr_time - prev_time)
    else:
        fps = 0
    prev_time = curr_time

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    fps_text = f'FPS : {int(fps)}'
    res_text = f'{int(real_w)} X {int(real_h)}'
    if show_info :
        cv2.putText(img,fps_text,(50,50), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,255,0), 2)
        cv2.putText(img,res_text,(50,100), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,255,0), 2)

    face_list = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=7, minSize=(150, 150))
    # 검출된 물체를 표시
    # scaleFactor = 이미지를 몇 배씩 줄여가면 탐색할지 지정하는 값
    # 얼굴은 1.1~1.3 사이에서 검출하면 적당함, 미소는 1.5~1.8이 적당
    # minNeughbors : 후보영역을 진짜 객체로 인정하려면 주변에서 몇 번이나 겹쳐 검출해야 하는지 횟수를 지정
    for(x,y,w,h) in face_list:
        cv2.rectangle(img,(x,y),(x+w, y+h),(0,255,0), 2)
        face_roi_gray = gray[y:y+h, x:x+w]
        face_roi_color = img[y:y+h, x:x+w]
        smiles = smile_cascade.detectMultiScale(face_roi_gray, scaleFactor=1.8, minNeighbors=30)
        for (sx, sy, sw, sh) in smiles:
            # 웃음 사각형 표시 (노란색)
            cv2.rectangle(face_roi_color, (sx, sy), (sx+sw, sy+sh), (0, 255, 255), 2)

    cv2.imshow('webcam', img)




   
    key = cv2.waitKey(1)
    if key == ord('q'): break
    elif key == ord('s'):
        filename = time.strftime('snapshot_%y%m%d%H%M%s.jpg')#day6 밑에 만들어짐
        cv2.imwrite(filename, img)
        print(f'{filename} 스냅샷 저장')
    elif key == 9:# 탭 아스키 코드번호
        show_info = not show_info
        print('탭')

cap.release() 
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### 얼굴 모자이크

```python
import cv2
import time


face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')


cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise RuntimeError('웹캠 오픈 실패')

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

prev_time = 0
show_info = False 

# 실제 해상도 읽기
real_w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
real_h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

while True:
    ret, img = cap.read()
    if not ret: break
    curr_time = time.time()

    if prev_time != 0 :
        fps = 1/(curr_time - prev_time)
    else:
        fps = 0
    prev_time = curr_time

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    fps_text = f'FPS : {int(fps)}'
    res_text = f'{int(real_w)} X {int(real_h)}'
    if show_info :
        cv2.putText(img,fps_text,(50,50), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,255,0), 2)
        cv2.putText(img,res_text,(50,100), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,255,0), 2)

    face_list = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(150, 150))
    for(x,y,w,h) in face_list:
        #cv2.rectangle(img,(x,y),(x+w, y+h),(0,255,0), 2)
        face_roi_gray = img[y:y+h, x:x+w]
        small = cv2.resize(face_roi_gray,(w//5, h//15), interpolation=cv2.INTER_LINEAR)
        mosaic = cv2.resize(small, (w,h), interpolation=cv2.INTER_NEAREST)

        img[y:y+h, x:x+w] = mosaic
        
    cv2.imshow('webcam', img)




   
    key = cv2.waitKey(1)
    if key == ord('q'): break
    elif key == ord('s'):
        filename = time.strftime('snapshot_%y%m%d%H%M%s.jpg')#day6 밑에 만들어짐
        cv2.imwrite(filename, img)
        print(f'{filename} 스냅샷 저장')
    elif key == 9:# 탭 아스키 코드번호
        show_info = not show_info
        print('탭')

cap.release() 
cv2.destroyAllWindows()
cv2.waitKey(1)
```

#### QR코드 리더


```python
# qr코드 검출기 
import cv2
import webbrowser # 웹 브라우저용 모듈

qr_detector = cv2.QRCodeDetector()

# 웹캠 
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

is_opened = False #이미 열었는지 여부
last_data = "" # 최종 데이터

while True:
    ret, img = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    data, points, _ = qr_detector.detectAndDecode(gray)

    if points is not None and data:
        points = points.astype(int)
    
        for i in range(len(points[0])):
            pt1 = tuple(points[0][i])
            pt2 = tuple(points[0][(i+1)%len(points[0])])
            cv2.line(img,pt1, pt2, (0,255,0), 2)


    cv2.putText(img,data,(50,50), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,255,0), 2)
    cv2.imshow('fdf',img)
    

    if (data.startswith('http://') or data.startswith('https://')) and (is_opened and last_data != data) == False:
        webbrowser.open(data)   
        is_opened = True
        last_data = data
    else:
        last_data = ""
        is_opened = False
        




    
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)

```
