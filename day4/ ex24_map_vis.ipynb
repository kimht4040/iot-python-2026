#### 위경도 특성
- 소수점 4자리 이상 작성해야 제대로된 위치를 잡을 수 있음

```python
import folium

#37.551192, 126.988165
```

```python
loc = [37.551192,126.988165]
m = folium.Map(location=loc, zoom_start=12) # 지도 객체 생성, m 변수 할당
# 마커생성후 지도 변수에 추가
folium.Marker(location=loc, tooltip='soeul', popup='soeul tower').add_to(m)
m
```

#### 맵 타일 문제
-  사용 주인 지도 타일 서버 접속 문제
-  기본 설정된 OpenStreetMap 타일 서버를 다른 서버로 지정하면 해결

```python
loc = [37.551192,126.988165]
m = folium.Map(location=loc, zoom_start=15, tiles='cartodbpositron')
# 마커생성후 지도 변수에 추가
folium.Marker(location=loc, tooltip='soeul', popup='soeul tower').add_to(m)
m
```
