- 한글 설정

```python
# 리소스 인수, 폰트 관리자, 리소스 포함

import platform
if platform.system() == 'Darwin': # 맥
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows': # 윈도우
    plt.rc('font', family='Malgun Gothic')
else: # 리눅스 (코랩 등)
    plt.rc('font', family='NanumGothic')
# - 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

#### 포인트 모양 변경



```python
plt.plot(x,y, marker = 'o')
plt.title('점수추이 ')
plt.ylabel('score')
plt.xlabel('시간')
plt.show()
```

#### 스타일지정
- seaborn은 차트 생성전에 스타일 지정

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 기존 설정을 초기화하고 다시 설정
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# Seaborn 스타일 적용 시 폰트 정보를 다시 명시
sns.set_style('whitegrid', {'font.family':['AppleGothic']})
```

```python
sns.set_style('whitegrid')# 스타일 지정
sns.set_theme(font = 'AppleGothic', rc={'axes.unicode_minus' : False}) # 한글 깨짐 방지
```

```python
import unicodedata

# 텍스트를 NFC 방식으로 합쳐주는 함수
def nfc(text):
    return unicodedata.normalize('NFC', text)
```

```python

plt.plot(x,y, marker = 'o')
plt.title(nfc('점수추이'))
plt.ylabel('score')
plt.xlabel(nfc('시간'))
plt.show()
```

#### 3D 그래프


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import unicodedata
import platform
from mpl_toolkits.mplot3d import Axes3D

if platform.system() == 'Darwin': # 맥
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows': # 윈도우
    plt.rc('font', family='Malgun Gothic')
else: # 리눅스 (코랩 등)
    plt.rc('font', family='NanumGothic')
# - 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 텍스트를 NFC 방식으로 합쳐주는 함수
def nfc(text):
    return unicodedata.normalize('NFC', text)
```

```python
x = [1,2,3,4]
y =[10, 20, 15, 30]
z = [5, 10 , 8, 12]

fig = plt.figure() # 새로운 차트 선언해서 변수할당
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x,y,z)

ax.set_xlabel('x축')
ax.set_ylabel('y축')
ax.set_zlabel('z축')
```

```python
# 3d surface
x = np.linspace(-5, 5, 50) # -5에서5사이를 50개로 쪼갬
y = np.linspace(-5, 5, 50)

x, y = np.meshgrid(x,y)
z = x**2+ y**2
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z)
plt.show()
```
