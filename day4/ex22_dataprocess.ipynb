### 간단 예제

```python
data = {
    'name' : ['유고', '요미', '철수', '예성'],
    'score':[80,90,70,60] 
}

df = pd.DataFrame(data)
df
```

```python
# csv로 쓰기
df.to_csv('score.csv', encoding='utf-8')
```

```python
df.to_excel('score.xlsx')

#평균이상만 추출
df[df['score'] >= df['score'].mean()]
```

```python
#합격 여부
df['result'] = ['pass'if s >= 70 else 'fail' for s in df['score']]
df
```
