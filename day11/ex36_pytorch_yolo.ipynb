```python
# OpenCV로 동영상을 실시간 재생하면서 YOLO 적용
import cv2
import time
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
video_path = '/Users/kimht4040/Desktop/code/iot-python-2026/day11/KakaoTalk_Video_2026-05-22-10-45-40.mp4'

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise RuntimeError('동영상을 열 수 없습니다.')

prev_time = time.time()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    result = model.predict(frame, verbose=False)[0]
    annotated_frame = result.plot()

    curr_time = time.time()
    fps = 1 / (curr_time - prev_time)
    prev_time = curr_time

    cv2.putText(
        annotated_frame,
        f'FPS: {fps:.1f}',
        (10, 30),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2,
    )
    for box in result[0].boxes:
        cls_id = int(box.cls[0])
        label = model.names[cls_id]
        if label == 'person':
            detected = True
    
    if detected:
        cv2.putText(
            annotated_frame,
            "detected",
            (20,50),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0,0,255),
            2
        )
        
    

    cv2.imshow('YOLO Real-Time Detection', annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
